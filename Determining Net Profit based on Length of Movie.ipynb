<b><u>Creating a Boxplot of Mean Net Profits for Movies based on Duration of Movie</b></u>


```python
#import csv data frames
import pandas as pd
df1 = pd.read_csv('bom.movie_gross.csv.gz')
df2 = pd.read_csv('imdb.name.basics.csv.gz')
df3 = pd.read_csv('imdb.title.akas.csv.gz')
df4 = pd.read_csv('imdb.title.basics.csv.gz')
df5 = pd.read_csv('imdb.title.crew.csv.gz')
df6 = pd.read_csv('imdb.title.principals.csv.gz')
df7 = pd.read_csv('imdb.title.ratings.csv.gz')
df10 = pd.read_csv('tmdb.movies.csv.gz', index_col = "Unnamed: 0")
df11 = pd.read_csv('tn.movie_budgets.csv.gz', index_col = "id")
```

```python
#display data frames
display(df1.head(), df2.head(), df3.head(), df4.head(), df5.head(), df6.head(), df7.head(), df10.head())
df11.head()
```

```python
#set index of df4 to primary title
df4.set_index('primary_title', inplace = True)
```

```python
#set index of df11 to movie
df11.set_index('movie', inplace = True)
```

```python
#join together df4 and df11 to connect budget and gross to movie durations
joined_df = df4.join(df11, how='inner')
```

```python
#display joined df
joined_df.head()
```

```python
joined_df.info()
```

```python
#check for null values
joined_df.isnull().sum()
```

```python
#check shape of joined df
joined_df.shape
```

```python
#remove rows with missing genres
joined_df = joined_df[joined_df['genres'].notna()]
```

```python
joined_df.isnull().sum()
```

```python
#check new shape
joined_df.shape
```

```python
#remove rows with missing runtimes
joined_df = joined_df[joined_df['runtime_minutes'].notna()]
```

```python
#check new shape
joined_df.shape
```

```python
#remove dollar sign and convert object data types (production_budget, domestic_gross and worldwide_gross) to integers
joined_df[joined_df.columns[6:]] = joined_df[joined_df.columns[6:]].replace('[\$,]', '', regex=True).astype(int)
```

```python
#check to see that changes went through
joined_df.info()
```

```python
#create column for net profit 
joined_df['net_profit'] = joined_df['worldwide_gross'] - joined_df['production_budget']
```

```python
#convert production budget to a more readable unit (production budget in millions)
joined_df['net_profit_by_millions'] = joined_df['net_profit'].map(lambda x: x/1000000)
joined_df.head()
```

```python
#Check descriptive statistics for runtime minutes to determine appropriate way to sort runtime
joined_df['runtime_minutes'].describe()
```

```python
#Use lambda function to break up movies into two distinct groups (short movies and long movies) based on runtime minutes
joined_df['duration_of_movie'] = joined_df.runtime_minutes.map(lambda x:"Short movie (under 98 minutes)" if x < 98 else "Long movie (over 98 minutes)")
```

```python
#import visualization tools
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
joined_df.head()
```

```python
#create boxplot of net profit per short movies and long movies
plt.figure(figsize=(10,10))
sns.boxplot(
    data=joined_df,
    x='length_of_movie',
    y='net_profit',
    color='green').set_title('Net Profit Based on Duration of Movie')
```
