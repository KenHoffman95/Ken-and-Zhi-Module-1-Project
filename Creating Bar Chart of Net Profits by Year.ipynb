<b><u>Creating a Bar Chart of Mean Net Profits for Movies by Year</b></u>


```python
#import pandas and csv files
import pandas as pd
df1 = pd.read_csv('bom.movie_gross.csv.gz')
df2 = pd.read_csv('imdb.name.basics.csv.gz')
df3 = pd.read_csv('imdb.title.akas.csv.gz')
df4 = pd.read_csv('imdb.title.basics.csv.gz')
df5 = pd.read_csv('imdb.title.crew.csv.gz')
df6 = pd.read_csv('imdb.title.principals.csv.gz')
df7 = pd.read_csv('imdb.title.ratings.csv.gz')
df10 = pd.read_csv('tmdb.movies.csv.gz', index_col = "Unnamed: 0")
df11 = pd.read_csv('tn.movie_budgets.csv.gz', index_col = "id")
```

```python
#display various csv files data
display(df1.head(), df2.head(), df3.head(), df4.head(), df5.head(), df6.head(), df7.head(), df10.head())
df11.head()
```

```python
#look at df11 info 
df11.info()
```

```python
#convert object data types to integers (production_budget, domestic_gross, and worldwide_gross)
df11[df11.columns[2:]] = df11[df11.columns[2:]].replace('[\$,]', '', regex=True).astype(int)
```

```python
#check to see that object types are now integers
df11.info()
```

```python
#create column for net profit
df11['net_profit'] = df11['worldwide_gross'] - df11['production_budget']
```

```python
#check to see that net profit column was succesfully added
df11.head()
```

```python
#convert net_profit to net_profit by millions for easier to read column
df11['net_profit_by_millions'] = df11['net_profit'].map(lambda x: x/1000000)
df11.head()
```

```python
#sort movies by net profit
df11.sort_values(by = 'net_profit', ascending = False).head(25)
```

```python
#check out descriptive statistics
df11.describe()
```

```python
#update date column so that it can be sorted
df11['release_date'] = pd.to_datetime(df11['release_date'])
```

```python
#sort values
df11.sort_values(by = 'release_date', ascending = False)
```

```python
#add column to just display year of release
df11['Year'] = df11['release_date'].dt.strftime('%Y')
```

```python
df11.head()
```

```python
#create new dataset that organizes movie sums by year (last 15 years minus the last two (incomplete data))
year_df = df11.groupby('Year').mean().tail(15)
year_df1 = year_df.drop(['2019', '2020'])
year_df1
```

```python
#import matplotlib to graph
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#set x data
x = year_df1.index
```

```python
#set y data
y = year_df1['net_profit_by_millions']
```

```python
#plot last 15 years of movies with complete data
plt.figure(figsize = (10, 6))
plt.xlabel('Year')
plt.ylabel('Net Profit (000,000s)')
plt.title('Mean Net Profit for Movies by Year')
plt.bar(x, y)
```
