<b><u>Creating a Data Visualization of Net Profits for Movies by Genre</b></u>

```python
#import csv files
import pandas as pd
df1 = pd.read_csv('bom.movie_gross.csv.gz')
df2 = pd.read_csv('imdb.name.basics.csv.gz')
df3 = pd.read_csv('imdb.title.akas.csv.gz')
df4 = pd.read_csv('imdb.title.basics.csv.gz')
df5 = pd.read_csv('imdb.title.crew.csv.gz')
df6 = pd.read_csv('imdb.title.principals.csv.gz')
df7 = pd.read_csv('imdb.title.ratings.csv.gz')
df10 = pd.read_csv('tmdb.movies.csv.gz', index_col = "Unnamed: 0")
df11 = pd.read_csv('tn.movie_budgets.csv.gz', index_col = "id")
```

```python
#explore dataframes
display(df1.head(), df2.head(), df3.head(), df4.head(), df5.head(), df6.head(), df7.head(), df10.head())
df11.head()
```

```python
#remove rows with missing genres
df4 = df4[df4['genres'].notna()]
```

```python
#Since there are multiple genres for some movies, split these genres up into distinct groups
df4[['main_genre','sub_genre', 'third_genre']] = df4.genres.apply(lambda x: pd.Series(str(x).split(","))) 
```

```python
#confirm that proper changes were made
df4.head()
```

```python
#set df4 index to primary title
df4.set_index('primary_title', inplace = True)
```

```python
#set df11 index to movie
df11.set_index('movie', inplace = True)
```

```python
#join df4 and df11 to create a dataframe with genres and production budget, gross domestic, and gross worldwide
joined_df = df4.join(df11, how='inner')
```

```python
joined_df.head()
```

```python
#determine whether there are missing values for genre
joined_df.isnull().sum()
```

```python
#look at shape of joined df
joined_df.shape
```

```python
#remove $ sign from production budget, domestic gross and worldwide gross and convert object data type to integers
joined_df[joined_df.columns[9:]] = joined_df[joined_df.columns[9:]].replace('[\$,]', '', regex=True).astype(int)
```

```python
#check to see that data types were succesfully changed to integers
joined_df.info()
```

```python
#create net profit column by subtracting the worldwide gross from the production budget
joined_df['net_profit'] = joined_df['worldwide_gross'] - joined_df['production_budget']
```

```python
#convert production budget to a more readable unit (production budget in millions)
joined_df['net_profit_by_millions'] = joined_df['net_profit'].map(lambda x: x/1000000)
joined_df.head()
```

```python
#import visualization tools
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#create boxplot with the net profits based on genre
plt.figure(figsize=(20,10))
sns.boxplot(
    data=joined_df,
    x='main_genre',
    y='net_profit',
    color='yellow').set_title('Net Profit per Genre')
```
