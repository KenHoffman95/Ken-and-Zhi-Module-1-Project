```python
#import csv data frames
import pandas as pd
df1 = pd.read_csv('bom.movie_gross.csv.gz')
df2 = pd.read_csv('imdb.name.basics.csv.gz')
df3 = pd.read_csv('imdb.title.akas.csv.gz')
df4 = pd.read_csv('imdb.title.basics.csv.gz')
df5 = pd.read_csv('imdb.title.crew.csv.gz')
df6 = pd.read_csv('imdb.title.principals.csv.gz')
df7 = pd.read_csv('imdb.title.ratings.csv.gz')
df10 = pd.read_csv('tmdb.movies.csv.gz', index_col = "Unnamed: 0")
df11 = pd.read_csv('tn.movie_budgets.csv.gz', index_col = "id")
```

```python
#display data frames
display(df1.head(), df2.head(), df3.head(), df4.head(), df5.head(), df6.head(), df7.head(), df10.head())
df11.head()
```

```python
df11.info()
```

```python
#remove $ sign and convert object data (production_budget, domestic_gross and worldwide_gross) to integers
df11[df11.columns[2:]] = df11[df11.columns[2:]].replace('[\$,]', '', regex=True).astype(int)
```

```python
df11.info()
```

```python
#create column for net profit 
df11['net_profit'] = df11['worldwide_gross'] - df11['production_budget']
```

```python
df11.head()
```

```python
#look at descriptive statistics to determine appropriate ranges for production_budget
df11.production_budget.describe()
```

```python
#since production budget descriptive statistics are in exponents, convert production budget to a more readable unit (production budget in millions)
df11['production_budget_by_millions'] = df11['production_budget'].map(lambda x: x/1000000)
df11.head()
```

```python
#look at descriptive statistics for updated production_budget column (production_budget_by_millions)
df11['production_budget_by_millions'].describe()
```

```python
#use lambda functions to create new column to organize movies by amount spent on production (very low (lowest 25%), low (25th to 50th percentile), high (50th to 75th percentile), very high (highest 25%))
df11['production_range'] = df11.production_budget_by_millions.map(lambda x:"very low" if x < 5 else "low" if x> 5 and x < 17 else "high" if x> 17 and x < 40 else "very high")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#create scatterplot to show the relationship between production budget and net profit
sns.set()
plt.figure(figsize = (10,10))
data = df11
sns.scatterplot(data['production_budget'], data['net_profit'], marker="x")
```

```python
#create boxplot to show relationship between production budget and net profit 
plt.figure(figsize=(10,10))
sns.boxplot(
    data=df11,
    x='production_range',
    y='net_profit',
    color='red')
```
